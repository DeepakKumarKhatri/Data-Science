## Tuple
It's just like list but it is immutable
- Ordered
- Unchangeable (immutable)
- Duplicacy Allowed

```python
t1 = (1,2,3,4,5)
t2 = (5,4,3,2,1)
print(t1 == t2)

l1 = [1,2,3,4,5]
l2 = [5,4,3,2,1]
print(l1 == l2)
```

```python
## add values in tuples using iteration
tup = ()
print(type(tup))

for i in range(5):
    tup+=(i,)
print(tup)

# We can add manually like this
tup+=6,
print(tup)
```

```python
# empty tuple
t1 = ()
print(t1)

# Single item tuple
t2 = (1) # wrong way
print(t2)
print(type(t2))

t2 = (1,) #correct way
print(t2)
print(type(t2))

t3 = (1,2,3,4)
print(t3)

t4 = (1,2,3.4,"Deepak",True,1+2j)
print(t4)

t5 = 1,2,3,4,5
print(t5)

#type conversion
t6 = tuple("Hello")
print(t6)
```

## Sets
- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

```python
## Creating Sets
s = set()
print(s)
print(type(s))
```

```python
# 1D set
s = {1,2,3}
print(s)

# 2d set
s = {1,2,3,{1,2}}
```

```python
# No duplicate values
s = {1,2,3,True,(1,2,3,4,5)}
print(s)
```

```python
# No mutable items
s = {1,2,3,4,[1,2,3]}
```

```python
s1 = {1,2,3,4,5}
s2 = {5,4,3,2,1}
print(s1 == s2)
```

```python
## Accessing items
s1 = {1,2,3,4,5}
s1[0]
```

```python
s1[-1]
```

```python
s1[0:]
```

```python
## Editing items
s1 = {1,2,3,4,5}
s1[0] = 78
```

```python
## Add items
print(s1)
s1.add(78)
print(s1)

## add multiple items in once

s1.update([5,6,7,8,9,10])
print(s1)
```

```python
## Deleting items
s = {1,2,3,4,5}
del s
# print(s)
del s[0]
```

```python
s = {1,2,3,4,5}
s.discard(4) # if item not found will not throw an error
print(s)

s.remove(2) # if item not found will throw an error
print(s)

s.pop()
print(s)

s.clear()
print(s)
```

### Set Operations

```python
s1 = {1,2,3,4,5}
s2 = {6,3,8,4,10}

## Union
print(s1 | s2)
print(s1.union(s2))

```

```python
## Intersection
print(s1 & s2)
print(s1.intersection(s2))

```

```python

## Difference
print(s1 - s2)
print(s1.difference(s2))

```

```python

## Symmetric Difference - leave common and print rest
print(s1 ^ s2)
print(s1.symmetric_difference(s2))
```

```python
# Membership
print(1 in s1)
print(s1 in s2)
```

```python
## Iteration
for i in s2:
    print(i)
```

```python
print(s1)
print(len(s1))
print(sum(s1))
print(min(s1))
print(max(s1))
print(sorted(s1,reverse=True))
```

```python
# isdisjoint
s1 = {1,2,3,4,5}
s2 = {6,3,8,4,10}

print(s1.isdisjoint(s2))

s1 = {1,2,3,4,5}
s2 = {6,7,8,9,10}

print(s1.isdisjoint(s2))
```

```python
s1 = {1,2,3,4,5}
s2 = {1,2}

print(s1.issubset(s2))
print(s2.issubset(s1))
```

```python
print(s1.issuperset(s2))
```

```python
# copy
s1 = {1,2,3}
s2 = s1.copy()

print(s1)
s1.add(4)
print(s1)
print(s2)
```

## Frozenset
Frozen set is just an immutable version of a python set object

```python
## creating a frozenset
fs = frozenset([1,2,3,4])
print(fs)
print(type(fs))
```

```python
fs1 = frozenset([1,2,3,4])
fs2 = frozenset([5,6,7,8])

print(fs1 | fs2)

## what works and what does not?
# Works -> all read functions
# Not Works -> write functions
```

```python
## 2D Frozenset

fs = frozenset([1,2,3,4,frozenset([1,2,3])])
print(fs)
```

## Dictionary Comprehension

```python
{i:i**2 for i in range(1,11)}
```

```python
distances = {'Islamabad':1500,'Karachi':8900,'Hyderabad':780}

{key:value*0.62 for (key,value) in distances.items()}
```

```python
# Using zip

days = ['Sunday','Monday','Tuesday','Wednesday']
temperature = [50,67.8,34.0,98.7]

{i:j for i,j in zip(days,temperature)}
```

```python
products = {'phone':10,'laptop':32,'charger':0,'mobile':0}

{key:value for (key,value) in products.items() if value>0}
```
